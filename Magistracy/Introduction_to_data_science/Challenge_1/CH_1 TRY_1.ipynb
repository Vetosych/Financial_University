```python
# выделяем признаки

col_1 = pd.DataFrame(data = file_train 
                            .groupby('Loan Status')['Loan Status'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_2 = pd.DataFrame(data = file_train 
                            .groupby('Current Loan Amount')['Current Loan Amount'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_3 = pd.DataFrame(data = file_train 
                            .groupby('Term')['Term'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_4 = pd.DataFrame(data = file_train 
                            .groupby('Credit Score')['Credit Score'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_5 = pd.DataFrame(data = file_train 
                            .groupby('Annual Income')['Annual Income'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_6 = pd.DataFrame(data = file_train 
                            .groupby('Years in current job')['Years in current job'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_7 = pd.DataFrame(data = file_train 
                            .groupby('Home Ownership')['Home Ownership'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_8 = pd.DataFrame(data = file_train 
                            .groupby('Purpose')['Purpose'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_9 = pd.DataFrame(data = file_train 
                            .groupby('Monthly Debt')['Monthly Debt'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_10 = pd.DataFrame(data = file_train 
                            .groupby('Years of Credit History')['Years of Credit History'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_11 = pd.DataFrame(data = file_train 
                            .groupby('Months since last delinquent')['Months since last delinquent'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_12 = pd.DataFrame(data = file_train 
                            .groupby('Number of Open Accounts')['Number of Open Accounts'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_13 = pd.DataFrame(data = file_train
                            .groupby('Number of Credit Problems')['Number of Credit Problems'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_14 = pd.DataFrame(data = file_train 
                            .groupby('Current Credit Balance')['Current Credit Balance'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_15 = pd.DataFrame(data = file_train 
                            .groupby('Maximum Open Credit')['Maximum Open Credit'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_16 = pd.DataFrame(data = file_train 
                            .groupby('Bankruptcies')['Bankruptcies'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
col_17 = pd.DataFrame(data = file_train 
                            .groupby('Tax Liens')['Tax Liens'].count()
                            .reset_index(name='cnt')
                            .sort_values('cnt', ascending = False)
                    )
```
